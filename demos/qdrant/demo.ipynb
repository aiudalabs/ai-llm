```python
profile = db.collection("profiles")
docs = profile.get()

for doc in docs:
    print(doc.to_dict())
```

```python
for doc in docs:
    print(doc.to_dict().keys())
```

```python
documents = []

for doc in docs:
    if len(doc.to_dict()["about"]) > 1:
        documents.append({
            "about": doc.to_dict()["about"],
            "displayName": doc.to_dict()["displayName"],
            "age": int(doc.to_dict()["age"]) if len(doc.to_dict()["age"]) > 0 else 0,
            "tags": doc.to_dict()["tags"],
            "service": doc.to_dict()["service"],
        })
```

```python
qdrant_client.recreate_collection(
    collection_name="yomap_profiles",
    vectors_config=models.VectorParams(
        size=encoder.get_sentence_embedding_dimension(),  # Vector size is defined by used model
        distance=models.Distance.COSINE,
    ),
)
```

```python
qdrant_client.upload_points(
    collection_name="yomap_profiles",
    points=[
        models.PointStruct(
            id=idx, vector=encoder.encode(doc['about']).tolist(), payload=doc
        )
        for idx, doc in enumerate(documents)
    ],
)
```

```python
from qdrant_client.http.models import Filter, FieldCondition, MatchValue
```

```python
search_term = "Necesito un taxi"

search_result = qdrant_client.search(
  collection_name = "yomap_profiles",
  query_vector = encoder.encode(search_term),
  query_filter=models.Filter(
        must=[
            models.FieldCondition(key="age", range=models.Range(gte=10)), 
            FieldCondition(key="service.text", match=MatchValue(value="transport"))
        ],
    ),
  limit = 5
)
search_result
```
